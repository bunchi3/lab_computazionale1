# IVP Euler's method - systems of differential equations

Solve the following IVPs using Euler's method with $n=1000$ steps. Plot the solution and its first derivative together on one plot, and plot the error in each component as functions of time on another.   

```julia
using Plots
using Markdown
using LaTeXStrings
using Printf
include("c:\\ALL\\Stefano\\Bicocca\\3terzo_anno\\lab_comp\\lab_computazionale1\\librerie\\diff_equations.jl")
```

```julia
n = 1000

f_a = t -> 1/5*sin(3t) + 2*cos(3t) + (1/5)*sin(2t)
f_a_der = t -> (3/5)*cos(3t) - 6*sin(3t) + (2/5)*cos(2t)
f_b = t -> exp(2.0*t) + exp(-2.0*t) - t
f_b_der = t -> -2.0*exp(-2.0*t) + 2.0*exp(2.0*t) - 1.0
f_c = t -> (3.0*t + 5.0/4.0)*exp(-2.0*t) + (t-1.0)/4.0
f_c_der = t -> (exp(-2.0*t))*(-6.0*t + 0.25*exp(2*t) + 0.5)

function f1_a(t, z1, z2)
    return z2
end 
function f2_a(t, z1, z2)
    return sin(2t) - 9 * z1
end
function f1_b(t, z1, z2)
    return z2
end
function f1_b(t, z1, z2)
    return 4*t + 4*z1
end
function f1_c(t, z1, z2)
    return z2
end
function f1_c(t, z1, z2)
    return t - 4*z1 - 4*z2
end


Fa = [f1_a, f2_a]
Fb = [f1_b, f2_a]
Fc = [f1_c, f2_a]

u0_a = [2, 1]
u0_b = [2, -1]
u0_c = [1, 3/4]

colors = ["#0F2080",  # blu scuro
          "#A95AA1",  # viola
          "#F5793A",  # arancione
          "#85C0F9"]  # azzurro
```

**(a)** $y''+ 9y = \sin(2t)$, $\ 0< t< 2\pi$, $\ y(0) = 2$, $\ y'(0) = 1$; 

$\quad$ $\hat{y}(t) = (1/5) \sin(3t) + 2 \cos (3t)+  (1/5) \sin (2t)$

$$
\begin{cases}
z_1' = z_2 \\
z_2' = \sin(2t) - 9z_1
\end{cases}
$$

where $z_1 = y$ and $z_2 = y'$

```julia
a, b = 0, 2*pi
t = [i for i in a:abs(b-a)/n:b]
Y = euler_ode(Fa, u0_a, n, a, b)

x_trig_index = [L"0", L"\frac{\pi}{4}", L"\frac{\pi}{2}", L"\frac{3\pi}{4}", L"\pi", L"\frac{5\pi}{4}", L"\frac{3\pi}{2}", L"\frac{7\pi}{4}", L"2\pi"]
fig = plot(figsize=(800, 600),
           xlabel=L"t", ylabel=L"u(t)",
           framestyle=:box,
           grid=true, gridalpha=0.5,
           xticks=([i*pi/2 for i in 0:0.5:4], x_trig_index),
           legend = :topleft
          )

plot!(fig, t, f_a.(t),
      label = L"\hat{u}(t)",
      color= colors[1],
      lw = 4.0,
     )

plot!(fig, t, f_a_der.(t),
      label = L"\hat{u}'(t)",
      color= colors[2],
      lw = 4.0,
     )

plot!(fig, t, Y[1],
      label = L"u_{approx}(t)",
      color= colors[3],
      linestyle = :dash,
      lw = 2.0,
     )

plot!(fig, t, Y[2],
      label = L"u'_{approx}(t)",
      color= colors[4],
      linestyle = :dash,
      lw = 2.0,
     )

display(fig)
```

```julia
err_1 = abs.(f_a.(t) .- Y[1])
err_2 = abs.(f_a_der.(t) .- Y[2])

#= err1_max = []
for i in 0:1:7
      push!(err1_max, findmax(err_1[Int(i*(length(err_1)/7)):Int((i+1)*(length(err_1)/7))]))
end =#

fig = plot(figsize=(800, 600),
           xlabel=L"t", ylabel=L"|u-\hat{u}\ |",
           framestyle=:box,
           grid=true, gridalpha=0.5,
           xticks=([i*pi/2 for i in 0:0.5:4], x_trig_index), yticks=0:0.125:1.125,
           yformatter = y -> @sprintf("%.2f", y),
           legend = :topleft
          )

plot!(fig, t, err_1,
      label = L"|u-\hat{u}\ |",
      color= :navy,
      lw = 2.0,
     )
#= scatter!(fig, t, err1_max) =#
plot!(fig, t, err_2,
      label = L"|u'-\hat{u}'|",
      color= :darkorange,
      lw = 2.0,
     )
display(fig)
```

**(b)** $y''- 4y = 4t$, $\: 0< t< 1.5$, $\: y(0) = 2$, $\: y'(0) = -1$;
    
$\quad$ $\hat{y}(t) = e^{2t} + e^{-2t}-t$

$$
\begin{cases}
z_1' = z_2 \\
z_2' = 4t + 4z_1
\end{cases}
$$

where $z_1 = y$ and $z_2 = y'$

```julia
a, b = 0, 1.5
t = [i for i in a:abs(b-a)/n:b]
Y = euler_ode(Fb, u0_b, n, a, b)

fig = plot(figsize=(800, 600),
           xlabel=L"t", ylabel=L"u(t)",
           framestyle=:box,
           grid=true, gridalpha=0.5,
           xticks=[i for i in a:0.25:b], yticks = -10^4:5*10^2:10^3,
           yformatter = y -> @sprintf("%.2e", y),
           legend = :topleft
          )

plot!(fig, t, f_b.(t),
      label = L"\hat{u}(t)",
      color= colors[1],
      lw = 2.0,
     )

plot!(fig, t, f_b_der.(t),
      label = L"\hat{u}'(t)",
      color= colors[2],
      lw = 2.0,
     )

plot!(fig, t, Y[1],
      label = L"u_{approx}(t)",
      color= colors[3],
      linestyle = :dash,
      lw = 2.0,
     )

plot!(fig, t, Y[2],
      label = L"u'_{approx}(t)",
      color= colors[4],
      linestyle = :dash,
      lw = 2.0,
     )

display(fig)
```

```julia
err_1 = abs.(f_b.(t) .- Y[1])
err_2 = abs.(f_b_der.(t) .- Y[2])

fig = plot(figsize=(800, 600),
           xlabel=L"t", ylabel=L"Errors",
           framestyle=:box,
           grid=true, gridalpha=0.5,
           xticks=[i for i in a:0.25:b], yticks = 0:5*10^2:10^4,
           yformatter = y -> @sprintf("%.2e", y),
           legend = :topleft
          )

plot!(fig, t, err_1,
      label = L"|u-\hat{u}\ |",
      color= :navy,
      lw = 2.0,
     )
plot!(fig, t, err_2,
      label = L"|u'-\hat{u}'|",
      color= :darkorange,
      lw = 2.0,
     )
display(fig)
```

**(c)** $y''+ 4y'+ 4y = t$, $\: 0< t< 4$, $\: y(0) = 1$,$\: y'(0) = 3/4$;
    
$\quad$ $\hat{y}(t) = (3t+5/4)e^{-2t} + (t-1)/4$

$$
\begin{cases}
z_1' = z_2 \\
z_2' = t - 4z_1 - 4z_2
\end{cases}
$$

where $z_1 = y$ and $z_2 = y'$

```julia
a, b = 0, 4.0
t = [i for i in a:abs(b-a)/n:b]
Y = euler_ode(Fc, u0_c, n, a, b)

fig = plot(figsize=(800, 600),
           xlabel=L"t", ylabel=L"u(t)",
           framestyle=:box,
           grid=true, gridalpha=0.5,
           xticks=[i for i in a:0.5:b], yticks=[i for i in -10^7:10^6:10^7],
           yformatter = y -> @sprintf("%.2e", y),
           legend = :topleft
          )

plot!(fig, t, f_c.(t),
      label = L"\hat{u}(t)",
      color= colors[1],
      lw = 4.0,
     )

plot!(fig, t, f_c_der.(t),
      label = L"\hat{u}'(t)",
      color= colors[2],
      lw = 2.0,
     )

plot!(fig, t, Y[1],
      label = L"u_{approx}(t)",
      color= colors[3],
      linestyle = :dash,
      lw = 2.0,
     )

plot!(fig, t, Y[2],
      label = L"u'_{approx}(t)",
      color= colors[4],
      linestyle = :dash,
      lw = 2.0,
     )

display(fig)
```

```julia
err_1 = abs.(f_c.(t) .- Y[1])
err_2 = abs.(f_c_der.(t) .- Y[2])

fig = plot(figsize=(800, 600),
           xlabel=L"t", ylabel=L"Errors",
           framestyle=:box,
           grid=true, gridalpha=0.5,
           xticks=[i for i in a:0.5:b], yticks=0:10^6:10^7,
           yformatter = y -> @sprintf("%.2e", y),
           legend = :topleft
          )

plot!(fig, t, err_1,
      label = L"|u-\hat{u}\ |",
      color= :navy,
      lw = 2.0,
     )
plot!(fig, t, err_2,
      label = L"|u'-\hat{u}'|",
      color= :darkorange,
      lw = 2.0,
     )
display(fig)
```
